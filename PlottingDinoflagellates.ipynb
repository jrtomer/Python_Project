```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt




objects = ('High Light', 'Low Light')
x = ('Low Light'), ('Low_Light')
y = [0.2967, 0.1963]
y_pos = np.arange(len(objects))
growth_rates = [0.2967,0.1963]
yerr = [0, 0.1867]

plt.bar(y_pos, growth_rates, align='center', alpha=0.5)
plt.errorbar(x, y, yerr, linestyle='none')
plt.xticks(y_pos, objects)
plt.xlabel('Treatment Groups')
plt.ylabel('Average Intrinsic Growth Rate')
plt.title('Comparing Growth Rates Across Treatment Groups')


plt.show(plt.savefig("LabreportBargraph.jpg"))

```
